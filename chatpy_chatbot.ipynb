```python
# Enhanced Rule-Based Chatbot with IST Time and Smart Keyword Detection
# Author: Akshaya Manoj
# AI Course Project (DA-III: Chatbot Development)

import random
import datetime
import pytz  # for timezone handling

def chatbot_response(user_input):
    user_input = user_input.lower()

    greetings = ["hi", "hello", "hey", "good morning", "good evening"]
    farewells = ["bye", "goodbye", "see you", "exit"]

    # Handle greetings
    if any(word in user_input for word in greetings):
        return random.choice(["Hello there!", "Hey! How can I help you?", "Hi! What brings you here today?"])

    # Handle farewells
    elif any(word in user_input for word in farewells):
        return random.choice(["Goodbye! ðŸ‘‹", "See you soon!", "Take care!"])

    # Handle time and date queries (in IST)
    elif "time" in user_input:
        ist = pytz.timezone('Asia/Kolkata')
        current_time = datetime.datetime.now(ist).strftime("%I:%M %p")
        return f"The current time in IST is {current_time}."

    elif "date" in user_input:
        ist = pytz.timezone('Asia/Kolkata')
        current_date = datetime.datetime.now(ist).strftime("%d %B %Y")
        return f"Today's date (IST) is {current_date}."

    # Handle questions about the chatbot
    elif "your name" in user_input or "ur name" in user_input:
        return "I'm ChatPy â€” your simple AI chatbot assistant!"

    elif "who created" in user_input or "made you" in user_input or "your creator" in user_input:
        return "I was created by Akshaya Manoj as part of an AI course project. ðŸ˜Š"

    elif "how were you made" in user_input or "how are you made" in user_input:
        return "I was programmed in Python using basic rule-based logic!"

    elif "what were you made for" in user_input or "your purpose" in user_input:
        return "I was made to help demonstrate chatbot development in an AI course project."

    elif "how are you" in user_input:
        return "I'm doing great! Thanks for asking. How about you?"

    elif "thank" in user_input:
        return random.choice(["You're welcome!", "No problem!", "Happy to help!"])

    # Fallback (no keyword match)
    else:
        return random.choice([
            "Sorry, I didnâ€™t quite catch that. Could you rephrase?",
            "Hmm, thatâ€™s interesting! Tell me more.",
            "I'm not sure about that, but Iâ€™d love to learn!"
        ])

def start_chat():
    print("ðŸ¤– ChatPy: Hello! Type 'bye' to exit.\n")
    while True:
        user_input = input("You: ")
        if any(word in user_input.lower() for word in ["bye", "goodbye", "exit"]):
            print("ðŸ¤– ChatPy: Goodbye! ðŸ‘‹ Have a great day!")
            break
        response = chatbot_response(user_input)
        print("ðŸ¤– ChatPy:", response)

start_chat()


```
